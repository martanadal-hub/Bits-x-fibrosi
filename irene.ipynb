```python
import pandas as pd

# Replace 'your_dataset.xlsx' with the path to your Excel file
file_path = 'BBDDFPFcodificada.xlsx'

# Read the Excel file with the first row as header
df = pd.read_excel(file_path, header=1)

# Display the first few rows of the dataframe
df.head()
```

```python
duplicated_pacients = df[df['NHC'].duplicated()]
duplicated_pacients
```

```python
df = df.drop(columns=['Pacient'])
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Eliminar columnas con más de 150 valores faltantes
df = df.dropna(thresh=150, axis=1)

# Mostrar las primeras filas del dataframe actualizado
df.head()
```

```python
import matplotlib.pyplot as plt

# Select numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plot each numeric column
for column in numeric_columns:
    plt.figure()
    df[column].plot(kind='hist', title=column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
