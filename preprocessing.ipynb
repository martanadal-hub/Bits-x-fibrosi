```python
import pandas as pd

# Carregar la base de dades
df = pd.read_excel('BBDDFPFcodificada.xlsx', header=1)

# Mostrar les primeres files del dataframe
df.head()
```

```python

df = df.drop(columns=['Pacient'])
df.head()
```

```python
duplicados = df[df['NHC'].duplicated(keep=False)]

# Mostrar pacientes duplicados
if not duplicados.empty:
    print("Pacientes duplicados encontrados:")
    print(duplicados)
else:
    print("No hay pacientes duplicados.")
```

```python
df['Patient_ID'] = df['NHC']
df.head()
```

```python
df.columns
```

```python
df['New_Death'] = df.apply(lambda row: 'Yes' if row['Death'] == 'No' and row['Transplantation date'] != 'No transplantation' else row['Death'], axis=1)
df.head()
```

```python
# Crear un nuevo DataFrame con las columnas deseadas
df_visualization = df[['New_Death', 'Death', 'Transplantation date']]

# Mostrar el nuevo DataFrame
print(df_visualization)
```

```python
df['progressio'] = df['FVC (%) at diagnosis'] - df['FVC (%) 1 year after diagnosis']
df.head()
```

```python
missing_progressio = df['progressio'].isnull().sum()
print(f"NÃºmero de valores faltantes en 'progressio': {missing_progressio}")
```

```python
df['si/no progressio'] = df['progressio'].apply(lambda x: 'yes' if x > 5 else 'no')
df.head()
```

```python
progressio_counts = df['si/no progressio'].value_counts()
print(progressio_counts)
```

```python
# Filtrar los datos donde ALP es igual a 1
filtered_df = df[df['Comorbidities'] == 0]

# Seleccionar las columnas 'New_Death' y 'si/no progressio'
result = filtered_df[['New_Death', 'si/no progressio']]

# Contar el nÃºmero de 'Yes' y 'No' en la columna 'New_Death'
counts_new_death = result['New_Death'].value_counts()

# Contar el nÃºmero de 'yes' y 'no' en la columna 'si/no progressio'
counts_progressio = result['si/no progressio'].value_counts()

# Mostrar el resultado
print("Counts for New_Death:")
print(counts_new_death)
print("\nCounts for si/no progressio:")
print(counts_progressio)
```
